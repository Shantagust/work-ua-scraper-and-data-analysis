# Import library

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Parse data from csv file

```python
df = pd.read_csv(os.path.join("..", "skills.csv"))
df
```

# Generate Rating by required skills

```python
skills_count = df.groupby("skill").size().reset_index(name="count")
sorted_skills = skills_count.sort_values(by="count", ascending=False).head(25)
plt.figure(figsize=(10, 10))
plt.title("Top required 25 Skills")
plt.bar(sorted_skills["skill"], sorted_skills["count"])
plt.xticks(rotation=90)
```
